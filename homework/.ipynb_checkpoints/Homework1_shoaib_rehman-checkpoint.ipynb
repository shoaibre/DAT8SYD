```python
# for each occupation in 'users', count the number of occurrences

# for each occupation, calculate the mean age

# for each occupation, calculate the minimum and maximum ages

# for each combination of occupation and gender, calculate the mean age

# randomly sample a DataFrame

# detect duplicate users
```

```python
#STEP 1 : READ DATA INTO A DATA FRAME 'users'
import pandas as pd
users=pd.read_table('../data/u.user',header=None)
#ARRANGE DATA IN TABULAR FORMAT SEPARATED BY PIPE OPERATOR
col_names=['User_id','Age','Sex','Occupation','Zip_Code' ]
users=pd.read_table('../data/u.user', header=None, sep='|' ,names=col_names, index_col='User_id')
users.head()
```

```python
# for each occupation in 'users', count the number of occurrences
#by_occ.describe()
users.Occupation.value_counts()
```

```python
#for each occupation, calculate the mean age
#Group data by occupation
by_occ=users.groupby('Occupation')
import numpy as np
by_occ['Age'].agg(np.mean)
```

```python
#for each occupation, calculate the minimum and maximum ages
by_occ['Age'].agg([np.min ,np.max])
```

```python
# for each combination of occupation and gender, calculate the mean age
by_occ_gender=users.groupby(['Sex','Occupation'])
#by_occ_gender.describe()
by_occ_gender.agg(np.mean)
```
